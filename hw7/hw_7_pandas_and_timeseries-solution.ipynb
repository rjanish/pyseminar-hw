```python
import json

import numpy as np
import pandas as pd

%pylab inline
```

```python
# load GitHub issue reports
# strip off all but 'title', 'created_at', 'labels', 'closed_at', 'id', and login 'user' data
# add to pandas dataframe
columns_except_user = ['title', 'created_at', 'labels', 'closed_at', 'id']  
closed = json.load(open("closed.json", 'r'))
cleaned = [dict([(col, entry[col]) for col in columns_except_user] + 
                [("user", entry["user"]["login"])]) for entry in closed] 
issues = pd.DataFrame(cleaned)
```

```python
# remove duplicate id's
issues = issues.drop_duplicates(cols="id")
```

```python
# convert 'created_at' and 'closed_at' to datetime objects
for date_field in ['created_at', 'closed_at']:
    issues[date_field] = pd.to_datetime(issues[date_field])
```

```python
# number of issues created per month
users = issues['user']
users.index=issues["created_at"]
issues_per_month = users.resample("M", how='count')
fig, ax = subplots(1,1)
issues_per_month.plot()
ax.set_title("Number of GitHub Issues per Month")
ax.set_ylabel("Count")

# number of unique users posting issues per month
def unique_count(series):
    '''compute the number of unique entries in a pandas series'''
    return len(np.unique(series))

unique_users_per_month = users.resample("M", how=unique_count)
fig, ax = subplots(1,1)
unique_users_per_month.plot()
ax.set_title("Number of Users Posting GitHub Issues per Month")
ax.set_ylabel("Count")
```

```python
 # plot of length of time to close issue
elapsed = (issues["closed_at"] - issues["created_at"])
elapsed = pd.Series(elapsed.values/np.timedelta64(1, 'D'), index=issues["created_at"]) # convert to days
elapsed_per_month = elapsed.resample('M', how='mean')
fig, ax = subplots(1,1)
elapsed_per_month.plot()
ax.set_title("Time Until Issue Closure by Month")
ax.set_ylabel("Days")

# table of length of time to close issue and number of issues
issue_table = pd.DataFrame({"mean_days_until_closure":elapsed_per_month, "number_of_issues":issues_per_month},
                           index=elapsed_per_month.index)

issue_table
```

```python
# create dataframe called 'comments' to hold all comments on all issues with unique id's
comments = []
for entry in closed:
    for comment in entry["comments"]:
        comments.append({"id":int(entry["id"]), "created":comment["created"], 
                         "author":comment["author"], "text":comment["text"]})
comments = pd.DataFrame(comments)
comments = comments.drop_duplicates(cols="id")
comments['created'] = pd.to_datetime(comments['created'].apply(int), unit='ms') # 'created' is initially in ms since unix epoch

comments.head()
```

```python
# monthly table of number comments, chattiest user, percentage comments by chattiest, and number unique commenters
authors = pd.Series(comments["author"])
authors.index = comments["created"]

def series_mode(series):
    '''compute the most comment element of a pandas Series'''
    if len(series) > 0:
        return series.value_counts().index[0]
    else:
        return nan 
    
def series_fraction(series, value):
    '''compute the fraction of values in a pandas' Series that are equal to the passed value'''
    if len(series) > 0:
        return (series == value).sum()/float(len(series))
    else:
        return nan
    
comments_per_month = authors.resample('M', how='count')
chattiest_per_month = authors.resample('M', how=series_mode)
percent_by_chattiest_per_month = authors.resample("M", how=lambda s:series_fraction(s, series_mode(s))*100.0)
unique_commenters_per_month = authors.resample('M', how=unique_count)

comments_table = pd.DataFrame({"number_of_comments":comments_per_month, "chattiest_user":chattiest_per_month, 
                               "percent_comments_by_chattiest_user":percent_by_chattiest_per_month, 
                               "number_of_unique_commenters":unique_commenters_per_month}, index=comments_per_month.index)
comments_table
```

```python
# create dataframe holding id and each individual label for every issue with unique id's 
labels = []
for entry in closed:
    if len(entry['labels']) == 0:
        labels.append((int(entry['id']), 'None'))
    else:
        for label in entry['labels']:
            labels += [(int(entry['id']), l) for l in label.values()]
labels = pd.DataFrame(labels, columns=['id', 'label'])
labels = labels.drop_duplicates(cols=['id','label'])

labels[:20]
```

```python
# merged labels table with the original table of issues data
# I've removed the 'labels' column from issues, as it seemed redundant
full_issues = pd.merge(labels, issues.drop('labels',1))
full_issues['days_to_closure'] = (full_issues['closed_at'] - full_issues['created_at']).values/np.timedelta64(1, 'D')

full_issues[:20]
```

```python
# create dataframe holding label type (no colors or urls) and closure time indexed by created_date for every unique issue id
label_types = []
for entry in closed:
    if len(entry['labels']) == 0:
        label_types.append((int(entry['id']), unicode('None')))
    else:
        for label in entry['labels']:
            label_types.append((int(entry['id']), label['name']))
label_types = pd.DataFrame(label_types, columns=['id', 'label'])
label_types = label_types.drop_duplicates(cols=['id','label'])
label_types = pd.merge(label_types, full_issues[['id', 'label', 'days_to_closure', 'created_at']], on=['id', 'label'])
label_types.index = label_types['created_at']
label_types = label_types.drop(['created_at', 'id'],1)

# group by label types type
grouped_by_type = label_types.groupby('label')

# compute monthly mean for Bug and Enchancment issues
for key, g in grouped_by_type:
    if key == 'Bug':
        bug_times = g        
    elif key == 'Enhancement':
        enhancement_times = g
enhancement_times_per_month = enhancement_times.drop('label',1).resample('M', how='mean')
enhancement_times_per_month = enhancement_times_per_month.rename(columns={'days_to_closure':"Enhancement"}) 
bug_times_per_month = bug_times.drop('label',1).resample('M', how='mean')
bug_times_per_month = bug_times_per_month.rename(columns={'days_to_closure':"Bug"})

fig, ax = subplots(1,1)
bug_times_per_month.plot(ax=ax)
enhancement_times_per_month.plot(ax=ax)
ax.set_title("Time to Close Bug and Enhancement Issues per Month")
ax.set_ylabel("Days")
ax.set_ylim(0,400)

# compute mean of each type
grouped_by_type.agg('mean')
```
